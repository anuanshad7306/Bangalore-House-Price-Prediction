# outer detection and handling

```python
df2 = df1.copy()
df2.info()
```

```python
#Define a function  to calculate IQR outliers
def detect_outliers(column):
  q1=column.quantile(0.25)
  q3=column.quantile(0.75)
  iqr=q3-q1
  lower_bound=q1-1.5*iqr
  upper_bound=q3+1.5*iqr
  return(column < lower_bound ) | (column > upper_bound)
```

```python
#apply the function to numeric columns
outliers_iqr=df2[['size','total_sqft','bath','balcony','price']].apply(detect_outliers)
```

```python
#View outliers
print('outliers detect using outliers')
print(outliers_iqr.sum())
```

```python
#filter out the rows with any outliers
df2=df2[~(outliers_iqr.any(axis=1))]
print("data after removing outliers (IQR):")
print(df2)
```

```python
import numpy as np
from scipy.stats import zscore
```

```python
# Compute Z-scores
z_scores = df2[[ 'size', 'total_sqft', 'bath',
       'balcony', 'price']].apply(zscore)
```

```python
# Define threshold
threshold = 3
```

```python
# Identify outliers
outliers_z = (z_scores.abs() > threshold)
```

```python
# View outliers
print("Outliers detected using Z-Score:")
print(outliers_z.sum())  # Count of outliers per column
```
